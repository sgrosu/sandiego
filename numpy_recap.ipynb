### Efficiency

```python
from timeit import Timer

size = 1000000
timeits = 1000
```

```python
arr = np.arange(size)
```

```python
timer = Timer('arr.sum()','from __main__ import arr')

timer.timeit(timeits)/timeits
```

```python
alist = list(range(size))
```

```python
timer_list = Timer('sum(alist)', 'from __main__ import alist')
timer_list.timeit(timeits)/timeits
```
