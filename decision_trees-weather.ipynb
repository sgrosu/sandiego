```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv('weather/daily_weather.csv')
```

```python
data.columns
```

```python
data.shape
```

```python
data[data.isnull().any(axis=1)].shape
```

```python
del data['number']
```

```python
data.head()
```

```python
data = data.dropna()
```

```python
data.shape
```

## Convert to a Classification Task

```python
clean_data = data.copy()
clean_data['high_humidity_label'] = (clean_data['relative_humidity_3pm'] > 24.99)*1
```

```python
clean_data.head()
```

### We are storing the high_humidity_label in variable y

```python
y = clean_data[['high_humidity_label']].copy()
#y
```

```python
morning_features = clean_data.columns.tolist()
morning_features = morning_features[:-3]
morning_features
```

```python
X = clean_data[morning_features].copy()
```

```python
X.columns
```

```python
y.columns
#type(y)
```

## Perform Test and Train Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=324)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
y_train.describe()
```

## Fit on train classifier

```python
humidity_classifier = DecisionTreeClassifier(max_leaf_nodes=10, random_state=0)
humidity_classifier.fit(X_train, y_train)
```

## Predict on test set

```python
predictions = humidity_classifier.predict(X_test)
```

```python
predictions[:10]
```

```python
y_test[:10]
```

## Measure Accuracy for Classifier

```python
accuracy_score(y_true = y_test, y_pred = predictions)
```
